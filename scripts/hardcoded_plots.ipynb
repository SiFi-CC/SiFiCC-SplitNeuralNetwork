### Distal Fall-Off Position Plot ###

```python
import numpy as np
import matplotlib.pyplot as plt

y = np.array([[0.480, 0.096, 0.476, 0.092],
              [4.678,8.482,4.584,4.686],
              [-3.128,1.064,-3.306,-3.122]])

y_err = np.array([[2.472,1.701,1.007,0.753],
                  [2.505,1.122,1.960,0.490],
                  [2.388,1.520,0.973,0.589]])

y1 = np.array([[-0.062,-0.182,-0.272,-0.302],
               [3.838,3.792, 3.900,3.940],
               [7.190,7.376,7.558,7.542],
               [-3.486,-3.444,-3.106,-3.172]])
y1_err = np.array([[1.867,1.132,0.743,0.558],
                   [1.991,1.435,0.811,0.520],
                   [1.755,1.473,0.798,0.495],
                   [2.517,1.465,0.925,0.497]])
```

```python

# increase font size
plt.rcParams.update({'font.size': 18})

def dfop_plot(data, data_err, labels, title=""):
    # determine number of Bgragg-peak positions
    k_bg = data.shape[0]
    
    assert len(labels) == k_bg
    
    plt.figure(figsize=(10, 8))
    # settings
    plt.title(title)
    plt.ylabel("Relative Distal fall-off position [mm]")
    # generate and set xticks
    tick_list = [i for i in range(k_bg)]
    tick_labels = ["BP at {}mm".format(i) for i in labels]
    plt.xticks(tick_list, tick_labels, rotation=0, ha="left")
    plt.yticks(np.arange(-12, 12,2.0))
    plt.ylim(-12,12)
    # plotting
    x = np.array([i for i in range(k_bg)])
    plt.hlines(y=0, xmin=-0.02, xmax=k_bg + 0.02, color="blue")       
    plt.errorbar(x+0.2, data[:,0], data_err[:,0], fmt=".", color="black", label=r"1x$10^8$ protons")
    plt.errorbar(x+0.4, data[:,1], data_err[:,1], fmt=".", color="red", label=r"2x$10^8$ protons")
    plt.errorbar(x+0.6, data[:,2], data_err[:,2], fmt=".", color="green", label=r"5x$10^8$ protons")
    plt.errorbar(x+0.8, data[:,3], data_err[:,3], fmt=".", color="blue", label=r"1x$10^9$ protons")
    # set cosmetics
    plt.grid()
    plt.legend(loc="upper left")
    plt.tight_layout()
    plt.show()
```

```python
dfop_plot(y, y_err, ["0", "5", "-5"], "EdgeConvResNetCluster")
dfop_plot(y1, y1_err, ["0", "5", "10", "-5"], "EdgeConvResNetSiPM")
```

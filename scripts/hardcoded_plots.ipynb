### Distal Fall-Off Position Plot ###

```python
# data

y1 = [9.044, 14.670]
y1_err = [3.157, 2.891]

y2 = [2.4968, 6.3068]
y2_err = [1.22168, 1.03904]
```

```python
import numpy as np
import matplotlib.pyplot as plt

# increase font size
plt.rcParams.update({'font.size': 18})

fig, axs = plt.subplots(1,2, figsize=(10, 6))
# set axs labels
axs[0].set_title("Cut-Based")
axs[1].set_title("DNN S1AX\nClas + filter")
axs[0].set_ylabel("Distal fall-off position [mm]")
axs[1].set_ylabel("Distal fall-off position [mm]")
# set axs ticks
axs[0].set_xticks([1,2],["BP at\n0 mm", "BP at\n5 mm"], rotation=0)
axs[1].set_xticks([1,2],["BP at\n0 mm", "BP at\n5 mm"], rotation=0)
axs[0].set_yticks(np.arange(0.0,22.0,2.0))
axs[1].set_yticks(np.arange(0.0,22.0,2.0))
# set axs limits
axs[0].set_xlim(0.5,2.5)
axs[1].set_xlim(0.5,2.5)
axs[0].set_ylim(-0.5, 20.0)
axs[1].set_ylim(-0.5, 20.0)
# plotting
axs[0].hlines(y=0, xmin=0.5, xmax=2.5, color="blue")
axs[1].hlines(y=0, xmin=0.5, xmax=2.5, color="blue")
axs[0].errorbar([1,2], y1, y1_err, fmt=".", color="black", label=r"4x$10^8$ protons")
axs[1].errorbar([1,2], y2, y2_err, fmt="X", color="red", label=r"4x$10^8$ protons")
# set cosmetics
axs[0].grid()
axs[1].grid()
axs[0].legend(loc="best")
axs[1].legend(loc="best")
plt.tight_layout()
plt.show()
```
